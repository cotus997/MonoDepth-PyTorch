```python
from typing import Tuple


def read_path_images(name_file: str) -> Tuple:
    with open(name_file) as f:
        left = [os.path.join("Data", os.path.abspath(line.rstrip().split()[0])) for line in f]
        right = [line.rstrip().split()[1] for line in f]
    return (left, right)
```
